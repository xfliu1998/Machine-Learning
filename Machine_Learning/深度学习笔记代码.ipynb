## 2.2回归logistics

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1./(1+np.exp(-x))

def relu(x):
    return np.where(x<0,0,x)

def plot_sigmoid():
    x = np.arange(-10, 10, 0.1)
    y = sigmoid(x)
    plt.plot(x,y)
    plt.show()
    
def plot_relu():
    x = np.arange(-10, 10, 0.1)
    y = relu(x)
    plt.plot(x,y)
    plt.show()
    
if __name__ == '__main__':
    plot_sigmoid()
    plot_relu()
```
